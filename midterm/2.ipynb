```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Загрузка данных (укажите правильный путь к вашему файлу)
data = pd.read_csv('gym_membership.csv')

# Membership type distribution
sns.countplot(data['abonoment_type'])
plt.title('Distribution of Membership Types')
plt.show()

# Box plot of visit frequency by membership type
sns.boxplot(x='abonoment_type', y='visit_per_week', data=data)
plt.title('Visit Frequency by Membership Type')
plt.show()
```
